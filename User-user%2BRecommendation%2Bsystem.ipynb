### About Dataset

Anonymous Ratings on jokes.

1. Ratings are real values ranging from -10.00 to +10.00 (the value "99" corresponds to "null" = "not rated").

2. One row per user

3. The first column gives the number of jokes rated by that user. The next 100 columns give the ratings for jokes 01 - 100.

### Read the dataset(jokes.csv)

Take care about the header in read_csv() as there are no column names given in the dataset.

### Find user-user similarity and item-item similarity

**Hint:** Use **cosine_similarity()**

1. Get the user-user similarity by having users as rows. `(ratings matrix)`
2. Get the item-item similarity by having items as rows. `(ratings.T)` transpose.

### 1.Consine Similarity

```python
cosine_users = linear_kernel(ratings_jokes, ratings_jokes)
```

### 2.Centered Cosine Similarity

Scale the ratings using StandardScaler

1. Pass the above data to StandardScaler().fit_transform() `consider it as ratings_diff` matrix

2. Use the above ratings_diff matrix to find 

    i. user-user similarity (cosine_similarity(ratings_diff))
    
    ii. item-item similarity (cosine_similarity(ratings_diff).T)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
ratings_diff = scaler.fit_transform(ratings_jokes)
```

```python
ratings_diff.shape
```

```python
cosine_users_scaled = linear_kernel(ratings_diff, ratings_diff)
cosine_users_scaled.shape
```

```python
cosine_items_scaled = linear_kernel(ratings_diff.T, ratings_diff.T)
cosine_items_scaled.shape
```
